```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={
    "Product_Category":["Electronics","Clothing","Furniture","Electronics","Clothing","Furniture"],
    "Sales":[2000,1500,800,550,2200,770],
    "Prices":[500,100,800,550,120,453],
    "Volume":[4,15,1,4,15,1],
    "Customer_demographics":["youth","Adults","seniors","youth","Adults","seniors"],
    "Time":["2024-01","2024-02","2024-01","2024-02","2024-01","2024-02"],
    "Discount":[10,15,5,12,10,7],
    "Region":["north","south","east","north","south","east"],
    "Return_Rate":[5,3,8,6,4,7]
}
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
grouped_sales=df.groupby(["Product_Category","Time"]).agg({
    "Sales":"sum",
    "Volume":"sum",
    "Discount":"mean",
    "Return_Rate":"mean"
}).reset_index()
```

```python
plt.figure(figsize=(10,6))
for category in df["Product_Category"].unique():
    category_data=grouped_sales[grouped_sales["Product_Category"]==category]
    plt.plot(category_data["Time"],category_data["Sales"],marker="o",label=category)
plt.title("Sales trends over time by product category")
plt.xlabel("Time")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
    
```

```python
plt.figure(figsize=(10,6))
for category in df["Product_Category"].unique():
    category_data=grouped_sales[grouped_sales["Product_Category"]==category]
    plt.bar(category_data["Time"],category_data["Volume"],alpha=0.7,label=category)
plt.title("sales trends over time by volume")
plt.xlabel("Time")
plt.ylabel("Volume")
plt.legend()
plt.show()
```
